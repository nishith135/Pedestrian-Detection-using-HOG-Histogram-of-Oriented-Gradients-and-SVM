```python
# ============================================
# UNIT 5 PROJECT: Pedestrian Detection using HOG + SVM
# ============================================

import cv2
import imutils
from google.colab import files
import matplotlib.pyplot as plt

# Step 1: Upload Image
print("ðŸ“¤ Please upload a street or pedestrian image (JPG/PNG):")
uploaded = files.upload()

if uploaded:
    image_path = list(uploaded.keys())[0]
else:
    raise Exception("No image uploaded. Please upload a valid image.")
```

```python
# Step 2: Initialize HOG + SVM Detector
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())

# Step 3: Read and preprocess the image
image = cv2.imread(image_path)
image = imutils.resize(image, width=min(600, image.shape[1]))

# Step 4: Detect pedestrians
(regions, _) = hog.detectMultiScale(image,
                                    winStride=(4, 4),
                                    padding=(4, 4),
                                    scale=1.05)
```

```python
# Step 5: Draw bounding boxes
for (x, y, w, h) in regions:
    cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 3)

# Step 6: Display results
plt.figure(figsize=(10, 6))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Pedestrian Detection using HOG + SVM")
plt.axis("off")
plt.show()

print(f"âœ… Detected {len(regions)} pedestrians in the image.")
```
